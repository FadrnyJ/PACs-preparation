```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gain1(alpha):
    return (2+np.abs(alpha)**2)/(1+np.abs(alpha)**2)

def gain2(alpha):
    return (6+6*np.abs(alpha)**2+np.abs(alpha)**4)/(2+4*np.abs(alpha)**2+np.abs(alpha)**4)

def var_x1(alpha):
    return (3+np.abs(alpha)**4)/(1+np.abs(alpha)**2)**2

def var_p1(alpha):
    return (3+np.abs(alpha)**2)/(1+np.abs(alpha)**2)

def var_x2(alpha):
    return (20+8*np.abs(alpha)**2+12*np.abs(alpha)**4+4*np.abs(alpha)**6+np.abs(alpha)**8)/(2+4*np.abs(alpha)**2+np.abs(alpha)**4)**2

def var_p2(alpha):
    return (10+8*np.abs(alpha)**2+np.abs(alpha)**4)/(2+4*np.abs(alpha)**2+np.abs(alpha)**4)
```

```python
alphas = np.arange(0,2.01,0.01)

data1 = np.loadtxt(r"Data_points\Fig4_ac.txt")
data2 = np.loadtxt(r"Data_points\Fig4_bd.txt")
```

# Plotting amplification gain

```python
plt_width=4
plt_ratio=0.75

fontsize = 22
markersize = 6
capsize=0
err_lw=1.4
lw=1.3
var_scale=1
DPI = 150

color1=(108/255,35/255,244/255,1)
color2=(230/255,52/255,81/255,1)
color3=(37/255,87/255,243/255,1)
```

```python
# Empty string will cause the script not to save the figure.
# If you want to save the figure, fill in the file name and the file type (i.e. .png, .svg)
filename = ""

spacs_alphas = data1[:,0]
spacs_gains = data1[:,1]
spacs_gains_std = data1[:,2]

plt.rcParams['text.usetex'] = True

c1=color3
%matplotlib inline
fig,ax = plt.subplots(figsize=(plt_width,plt_width*plt_ratio))
ax.plot(alphas,gain1(alphas),ls='-',c=c1,lw=lw)
ax.errorbar(spacs_alphas,spacs_gains, yerr = var_scale*spacs_gains_std,fmt ='ko',
            markersize=markersize,capsize=capsize,lw=err_lw)

ax.set_xlabel(r'$\vert \alpha \vert$',fontsize = fontsize+4)
ax.set_ylabel(r'$g$',fontsize = fontsize+4)
plt.xticks([0,0.5,1,1.5,2],fontsize=fontsize)
plt.yticks([1.2,1.4,1.6,1.8,2],fontsize=fontsize)
if filename != "":
    plt.savefig(filename, dpi=DPI, bbox_inches = 'tight')
```

```python
# Empty string will cause the script not to save the figure.
# If you want to save the figure, fill in the file name and the file type (i.e. .png, .svg)
filename = ""

tpacs_alphas = data2[:,0]
tpacs_gains = data2[:,1]
tpacs_gains_std = data2[:,2]

plt.rcParams['text.usetex'] = True

c1=color3
%matplotlib inline
fig,ax = plt.subplots(figsize=(plt_width,plt_width*plt_ratio))
ax.plot(alphas,gain2(alphas),ls='-',c=c1,lw=lw)
ax.errorbar(tpacs_alphas,tpacs_gains, yerr = var_scale*tpacs_gains_std,fmt ='ko',
            markersize=markersize,capsize=capsize,lw=err_lw)

ax.set_xlabel(r'$\vert \alpha \vert$',fontsize = fontsize+4)
ax.set_ylabel(r'$g$',fontsize = fontsize+4)
plt.xticks([0,0.5,1,1.5,2],fontsize=fontsize)
plt.yticks([1.5,2,2.5,3],fontsize=fontsize)
if filename != "":
    plt.savefig(filename, dpi=DPI, bbox_inches = 'tight')
```

# Ploting quadrature variances

```python
# Empty string will cause the script not to save the figure.
# If you want to save the figure, fill in the file name and the file type (i.e. .png, .svg)
filename = ""

spacs_alphas = data1[:,0]
spacs_det_var = data1[:,3]
spacs_det_var_std = data1[:,4]
spacs_varx = data1[:,5]
spacs_varx_std = data1[:,6]
spacs_varp = data1[:,7]
spacs_varp_std = data1[:,8]

plt.rcParams['text.usetex'] = True

c1=color2
c2=color3
c3=color1


%matplotlib inline
var_scale=1
fig,ax = plt.subplots(figsize=(plt_width,plt_width*plt_ratio))
plt.errorbar(spacs_alphas,spacs_det_var, yerr = var_scale*spacs_det_var_std,
             fmt ='o',color=c1,markersize=markersize,capsize=capsize,lw=err_lw)

plt.errorbar(spacs_alphas,spacs_varx, yerr = var_scale*spacs_varx_std,
             fmt ='s',color=c3,markersize=markersize,capsize=capsize,lw=err_lw)

plt.errorbar(spacs_alphas,spacs_varp, yerr = var_scale*spacs_varp_std,
             fmt ='^',color=c2,markersize=markersize,capsize=capsize,lw=err_lw)


plt.plot(alphas,2*gain1(alphas)**2 - 1,color=c1,ls='-',linewidth=lw)
plt.plot(alphas,var_p1(alphas),color=c2,ls='--',linewidth=lw)
plt.plot(alphas,var_x1(alphas),color=c3,ls='-.',linewidth=lw)


plt.plot(alphas,np.ones(len(alphas)),"k--",linewidth=lw)

plt.xlabel(r'$\vert \alpha \vert$',fontsize = fontsize+4)
plt.ylabel(r'$V_{x},V_{p}$',fontsize = fontsize+4)
plt.xticks([0,0.5,1,1.5,2],fontsize=fontsize)
plt.yticks(fontsize=fontsize)
ax.set_ylim(0,5)
if filename != "":
    plt.savefig(filename, dpi=DPI, bbox_inches = 'tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphas = np.arange(0,2.01,0.01)

data1 = np.loadtxt(r"Data_points\Fig6_a.txt")
data2 = np.loadtxt(r"Data_points\Fig6_b.txt")
data3 = np.loadtxt(r"Data_points\Fig6_c.txt")

prediction1 = np.loadtxt(r"Data_points\sw_F_n1_sr1.dat")
prediction2 = np.loadtxt(r"Data_points\srw_pacs_two.dat")
prediction3 = np.loadtxt(r"Data_points\srw_pacs_three.dat")
```

```python
plt_width=4
plt_ratio=0.75

fontsize = 22
markersize = 6
capsize=0
err_lw=1.4
lw=1.3
var_scale=1
DPI = 150

c3=(108/255,35/255,244/255,1)
c2=(230/255,52/255,81/255,1)
c1=(37/255,87/255,243/255,1)
```

```python
# Empty string will cause the script not to save the figure.
# If you want to save the figure, fill in the file name and the file type (i.e. .png, .svg)
filename = ""

spacs_amplitudes_str = data1[:,0]
fidelities1 = data1[:,1]
fid_err1 = data1[:,2]

plt.rcParams['text.usetex'] = True

%matplotlib inline
fig,ax = plt.subplots(figsize=(plt_width,plt_width*plt_ratio))

plt.plot(prediction1[:,0],prediction1[:,1],"-",color=c1,linewidth=lw)
plt.errorbar(spacs_amplitudes_str,fidelities1, yerr = var_scale*fid_err1,fmt ='ko',
             markersize=markersize,capsize=capsize,lw=err_lw)

plt.xlabel(r'$\vert \alpha \vert$',fontsize = fontsize+4)
plt.ylabel(r'$F_{th}$',fontsize = fontsize+4)


plt.xticks(fontsize=fontsize)
plt.yticks([0.4,0.6,0.8,1],fontsize=fontsize)
if filename != "":
    plt.savefig(filename, dpi=DPI, bbox_inches = 'tight')
```

# Plotting the inserts

```python
fock1 = np.loadtxt(r"Density_matrixes\fock1_DM.txt", dtype=np.complex128)
fock2 = np.loadtxt(r"Density_matrixes\fock2_DM.txt", dtype=np.complex128)
fock3 = np.loadtxt(r"Density_matrixes\fock3_DM.txt", dtype=np.complex128)

fock1 = np.diag(fock1)
fock2 = np.diag(fock2)
fock3 = np.diag(fock3)
```

```python
def insert_plot(data, dim=5, color = (179/255,179/255,179/255), lw=1.5, label_size=18, pad_x=3, pad_y=2, saveas="",dpi=200):

    data=data
    dim=dim

    dat=np.real(np.diag(data))[0:dim]
    dat_x = list(range(0,len(dat)))
    #print(np.sum(dat))

    col=color

    plt.rcParams['text.usetex'] = True

    lw=lw
    label_size = label_size
    pad_x,pad_y = pad_x,pad_y

    #%matplotlib notebook
    fig, ax = plt.subplots(figsize=(2*1.35*0.8,1.5*1.35*0.8))
    #fig, ax = plt.subplots(figsize=(2*1.35,1.5*1.35))
    ax.bar(dat_x, dat, edgecolor='k',linewidth=1.5, color=col)

    #ax.set_ylim=(0,1)

    ax.set_yticks((0,0.50,1))
    ax.set_xticks((0,1,2,3,4,5))
    #ax.set_xticks((0,2,4))

    ax.tick_params('x', labelsize=label_size, pad=pad_x)
    ax.tick_params('y', labelsize=label_size, pad=pad_y)

    ax.set_ylabel(r"$P_{m}$",fontsize=label_size)
    ax.set_xlabel(r"$m$",fontsize=label_size)

    #plt.setp(ax.spines.values(), lw=lw, color='r')

    plt.tick_params('x',direction='in', length=0, width=lw, colors='k')
    plt.tick_params('y',direction='in', length=6, width=lw, colors='k')
    if saveas !="":
        plt.savefig(filename,dpi=200)
```

```python
insert_plot(fock1)
```

```python
insert_plot(fock2)
```

```python
insert_plot(fock3)
```
